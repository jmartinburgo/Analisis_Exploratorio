```python
#Vamos a codificar las variables tipo object pero para eso tenemos que saber cuantos valores distintos tienen
print('Valores unicos de Gender:')
print(df_banco.Gender.unique())

print('Valores unicos de Geography:')
print(df_banco['Geography'].unique())

print('Valores unicos de Card Type:')
print(df_banco['Card Type'].unique())

```

```python
# Son pocas dimensiones entonces las codificaremos manualmente
codificacion_manual= {"Male":1,"Female":2}
df_banco["Gender"]=df_banco["Gender"].map(codificacion_manual)

cod_manual = {"Spain": 1, "France": 2, "Germany": 3}
df_banco["Geography"] = df_banco["Geography"].map(cod_manual)

c_manual = {"DIAMOND": 1, "GOLD": 2, "PLATINUM": 3, "SILVER": 4}
df_banco["Card Type"] = df_banco["Card Type"].map(c_manual)
```
