```python
#Analisis Univariado
#Variables Categoricas
fig, ax= plt.subplots(figsize=(15,3))

ax0=plt.subplot2grid((1,3),(0,0))
count=df_penguins['species'].value_counts()
plt.pie(x=count,labels=count.index,autopct='%.0f%%',startangle=90,colors=sns.color_palette('pastel'))
plt.title('Especies')

ax1=plt.subplot2grid((1,3),(0,1))
counts=df_penguins['sex'].value_counts()
plt.pie(x=counts,labels=counts.index,autopct='%.0f%%',startangle=90,colors=sns.color_palette('pastel'))
plt.title('Sexo')


ax2=plt.subplot2grid((1,3),(0,2))
count=df_penguins['island'].value_counts()
plt.pie(x=count,labels=count.index,autopct='%.0f%%',startangle=90,colors=sns.color_palette('pastel'))
plt.title('Isla')

plt.show()
```

```python
#Tendencias
fig, ax=plt.subplots(nrows=1,ncols=3, figsize=(15,4),sharex=True)

sns.countplot(data=df_penguins,x='year',hue='species',ax=ax[0])
ax[0].set_ylabel('Frecuencia')
ax[0].set_xlabel('Año')
sns.move_legend(ax[0],'lower center',bbox_to_anchor=(0.5,1),ncol=3,title='Especies',frameon=False)

sns.countplot(data=df_penguins,x='year',hue='sex',ax=ax[1])
ax[1].set_ylabel('Frecuencia')
ax[1].set_xlabel('Año')
sns.move_legend(ax[1], "lower center", bbox_to_anchor=(.5, 1), ncol=2, title='Sexo', frameon=False)

sns.countplot(data=df_penguins,x='year',hue='island',ax=ax[2])
ax[2].set_ylabel('Frecuencia')
ax[2].set_xlabel('Año')
sns.move_legend(ax[2], "lower center", bbox_to_anchor=(.5, 1), ncol=2, title='Sexo', frameon=False)

plt.show()
```

```python
#Analisis Multivariado
#Longitud del Pico [mm] vs Especies

#Histograma

var= 'bill_length_mm'
var_title='Longitud del pico [mm]'

fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10),sharex=False )


sns.histplot(data=df_adelie,x=df_adelie[var],ax=ax[0,0],label='Adelie')

sns.histplot(data=df_gentoo,x=var,ax=ax[0,0],label='Gentoo')

sns.histplot(data=df_chinstrap,x=var,ax=ax[0,0],label='Chinstrap')


ax[0,0].set_xlabel(var_title)
ax[0,0].set_ylabel('Frecuencia')
ax[0,0].legend()
sns.move_legend(ax[0,0],'lower center',bbox_to_anchor=(1,1),ncol=3)



#Curva de densidad

sns.kdeplot(data=df_adelie,x=var,ax=ax[0,1],label='Adelie',fill=True)

sns.kdeplot(data=df_gentoo,x=var,ax=ax[0,1],label='Gentoo',fill=True)

sns.kdeplot(data=df_chinstrap,x=var,ax=ax[0,1],label='Chinstrap',fill=True)

ax[0,1].set_xlabel(var_title)
ax[0,1].set_ylabel('Densidad')



#Boxplot

sns.boxplot(data=df_penguins, x='species',y=var,hue='species',ax=ax[1,0],width=.4)
ax[1,0].set_xlabel('Especies')
ax[1,0].set_ylabel(var_title)
ax[1,0].get_legend().remove()

sns.violinplot(data=df_penguins, x='species',y=var,hue='species',ax=ax[1,1],width=.4)
ax[1,1].set_xlabel('Especies')
ax[1,1].set_ylabel(var_title)
ax[1,1].get_legend().remove()


plt.show()
```
